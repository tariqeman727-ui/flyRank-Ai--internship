## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Target or Proxy

**Target:** Refresh Priority Score

The model predicts a refresh priority score for each webpage.

If a true priority label is not available, a proxy can be created using signals such as:
- Low organic traffic
- Declining clicks
- Old content age
- Low engagement
- Poor search ranking

Pages with these characteristics receive a higher refresh priority.
"""))
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Success Metric

**Success Metric:** NDCG (Normalized Discounted Cumulative Gain)

The model is successful if it ranks the most important pages at the top of the list.

A higher NDCG score means the model correctly prioritizes pages that should be refreshed first. This helps content teams focus on updates that are likely to have the greatest impact.
"""))
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
from IPython.display import Markdown, display
import pandas as pd

# Sample dataframe
df = pd.DataFrame({
    "page_url": [
        "/blog/seo-guide",
        "/blog/ai-tools",
        "/services/web-development"
    ],
    "organic_traffic": [1200, 450, 800],
    "clicks": [300, 90, 180],
    "content_age_months": [18, 24, 8]
})

display(df)

display(Markdown("""
### Unit of Analysis

Each row in this dataframe represents **one webpage**. The model will assign a refresh priority score to each page based on its performance metrics.
"""))
```
